# Delhi

```python
    for Pow in Pows:
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&near={}&limit={}&categoryId={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            cities[0],
            LIMIT,
            Pow)
        results[Pow] = requests.get(url).json()
        
        venues = json_normalize(results[Pow]['response']['groups'][0]['items'])
        df_venues[Pow] = venues[['venue.name', 'venue.location.address', 'venue.location.lat', 'venue.location.lng']]
        df_venues[Pow].columns = ['Name', 'Address', 'Lat', 'Lng']
        
        Pow_lat = np.mean([results[Pow]['response']['geocode']['geometry']['bounds']['ne']['lat'],
                        results[Pow]['response']['geocode']['geometry']['bounds']['sw']['lat']])
        Pow_lng = np.mean([results[Pow]['response']['geocode']['geometry']['bounds']['ne']['lng'],
                        results[Pow]['response']['geocode']['geometry']['bounds']['sw']['lng']])
        maps[cities[0]] = folium.Map(location=[Pow_lat, Pow_lng], zoom_start=11)
```

```python
        # adding Temple markers to map
        for lat, lng, label in zip(df_venues[Pows[0]]['Lat'], df_venues[Pows[0]]['Lng'], df_venues[Pows[0]]['Name']):
            label = folium.Popup(label, parse_html=True)
            folium.CircleMarker(
                [lat, lng],
                radius=5,
                popup=label,
                color='red',
                fill=True,
                fill_color='#F97E6D',
                fill_opacity=0.7,
                parse_html=False).add_to(maps[cities[0]])  
        print(f"Total number of Hindu Temples in {cities[0]} = ", results[Pows[0]]['response']['totalResults'])
    
        # adding Mosque markers to map
        for lat, lng, label in zip(df_venues[Pows[1]]['Lat'], df_venues[Pows[1]]['Lng'], df_venues[Pows[1]]['Name']):
            label = folium.Popup(label, parse_html=True)
            folium.CircleMarker(
                [lat, lng],
                radius=5,
                popup=label,
                color='green',
                fill=True,
                fill_color='#25EC59',
                fill_opacity=0.7,
                parse_html=False).add_to(maps[cities[0]])  
        print(f"Total number of Mosques in {cities[0]} = ", results[Pows[1]]['response']['totalResults'])
    
        # adding Church markers to map
        for lat, lng, label in zip(df_venues[Pows[2]]['Lat'], df_venues[Pows[2]]['Lng'], df_venues[Pows[2]]['Name']):
            label = folium.Popup(label, parse_html=True)
            folium.CircleMarker(
                [lat, lng],
                radius=5,
                popup=label,
                color='blue',
                fill=True,
                fill_color='#25A1EC',
                fill_opacity=0.7,
                parse_html=False).add_to(maps[cities[0]])  
        print(f"Total number of Churches in {cities[0]} = ", results[Pows[2]]['response']['totalResults'])
        maps[cities[0]]
```
